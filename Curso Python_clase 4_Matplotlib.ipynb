# Metodos Orientado a Objetos de Matplotlib
Matplotlib al ser una libreria completa, tiene tambien clases y metodos internos. Esto significa que se puede instanciar los objetos figuras y de ahi llamar metodos o atributos de estos.

## Introducción
La idea principal de utilizar el MOO es crear objetos figuras y de ahi llamar a sus metodos o atributos, esto se utiliza cuando se maneja multiples ploteos en graficos.

Primero se crea una instancia de figura. De ahi se le agrega ejes a la figura.

Quizas parezca mas complicado, pero la ventaja es que tendremos un control completo de donde ubicar nuestro plot y podemos facilmente agregar mas un eje a la figura:

```python
#Se crea un lienzo en blanco
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Figura Larga Eje 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

# Figura Larga Eje 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
#Crear figura, (lienzo vacio)
fig = plt.figure()

#Asignar ejes a la figura
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # izquierda, abajo, ancho, alto (range 0 a 1)
#axes=fig.add_subplot
# Plotear en el eje creado utilizando metodos
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title');
```


## subplots()
El comando subplots sirve tambien para organizar los ejes, se utiliza de igual manera que en Matlab.

```python
#Uso similar a plt.figure()
fig, axes = plt.subplots()

#Ahora se utiliza el objeto axes para agregar plots.
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

De ahi se puede especificar la cantidad de subplots a utilizar cuando se crea el objeto subplots()

```python
#Se crea un lienzo vacio de 1 x 2 
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
#El objeto Axes es un array de ejes donde se plotea
axes
```

Entonces se puede iterar sobre estos.

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
#mostrar la figura
fig
```

Un problema comun con matplot es la sobreposicion de los subplots o figuras. Podemos utilizar los metodos fig.tight_layout() o plt.tight_layout(), que automaticamente ajusta la posicion de los ejes en el lienzo.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig.tight_layout()
#plt.tight_layout()
```

### Leyendas
Se puede utilizar el argumento  label="label text" cuando se agrega texto a las etiquetas, y usando el metodo legend  sin argumentos para agegar la leyenda a la figura: 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

Se puede utilzar funciones especiales para ubicar el cuadro de la leyenda segun convenga.

```python
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner


# Most common to choose
ax.legend(loc=0) # Deja a matplotlib decidir la ubicacion ideal
fig
```

## Configurando colores, ancho de linea y tipos de linea
Matplotlib permite configurar basado en la misma sintaxis de Matlab

### Colores con Matlab como si fuesen sintaxis
Se puede deifnir los colores de la linea  y otros elementos graficos en un numero distinto de formas.
En Matlab se utiliza 'b' para azul, 'g' para verde, 'k' para negro, etc.
Tambien podemos usar la sintaxis para el tipo de linea.

```python
# MATLAB
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # linea azul con puntos
ax.plot(x, x**3, 'g--') # lineas intercaladas de color verde
```

### Uso de color con argumento color=
Podemos tambien definir el color por sus nombres en codigo HEX y tambien indicar la opacidad con el argumento alpha.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparantmedio-transparente
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

### Estilo de lineas
Para cambiar el estilo de linea se puede uilizar los argumentos 'ls' para el estilo y 'lw' para el ancho.

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

## Rango de graficos
Podemos limitar la longitud del eje a visualizar usando los metodos 'set_ylim' y 'set_xlim' en el objeto eje o usar axis.('tight') para automaticamente acercar los graficos.

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```

## Escalas logaritmicas

Es posible tambien cambiar la escala de los ejes, no solo limitarla. Usando el metodo 'set_xscale' y 'set_yscale'.

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
print(x)
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, 'r', x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```

## Ubicacion de marcas en ejes
Podemos decidir la ubicacion de los valores de los ejes utilizando los metodos 'set_xticks' y 'set_yticks', los cuales toman como valor una lista conteniendo los valores que seran ubicados en los ejes. Tambien podemos proveer una lista de string si se utiliza los metodos set_xticklabels y set_yticklabels.

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18); # use LaTeX formatted labels
```

## Malla de ejes
Con el metodo 'grid' podemos decidir activar o no la malla de fondo para guiarnos en la visualizacion del grafico.


```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# personalizable
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```
